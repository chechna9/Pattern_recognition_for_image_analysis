```python
import cv2 as cv
import  numpy as np
```

Convert to grayscale

```python
imagePath = "./img.png"
image = cv.imread(imagePath)
cv.imshow("Image", image)
# Convert the image to grayscale
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv.imshow("Gray Image", gray_image)
cv.waitKey(0)
```

### apply the thresholding and then find out the countours using cv2.findCountoures



```python
_,threshedImage = cv.threshold(gray_image,128,255,cv.THRESH_BINARY)
cv.imshow("Threshed Image", threshedImage)
cv.waitKey(0)
```

```python
contours, hierarchies = cv.findContours(threshedImage, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
```

```python
print(contours)
```

```python
def classify_shape(vertices):
    if vertices == 3:
        return "Triangle"
    elif vertices == 4:
        return "Rectangle"
    elif vertices == 5:
        return "Pentagon"
    else:
        return "Unknown"
```

```python
printedImage = image.copy()
for contour in contours:
    # Approximate the shape using cv.approxPolyDP
    epsilon = 0.04 * cv.arcLength(contour, True)
    approx = cv.approxPolyDP(contour, epsilon, True)

    # Draw the contour
    cv.drawContours(image, [approx], 0, (0, 255, 0), 2)

    # Find the center point of the shape
    M = cv.moments(contour)
    if M["m00"] != 0:
        cx = int(M["m10"] / M["m00"])
        cy = int(M["m01"] / M["m00"])
        center = (cx, cy)

        # countourArea
        area =  cv.contourArea(contour)
        
        # perimeter
        perimeter = cv.arcLength(contour, True)
        
        # roundness
        roundness = 4 * np.pi * area / perimeter**2
        # 
        # Classify the detected shape
        num_vertices = len(approx)
        shape_name = classify_shape(num_vertices)
        vertical_space = 30
        # Put the shape name at the center point of the shape
        cv.putText(printedImage, shape_name, (cx, cy), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
        cv.putText(printedImage, f'area {area}', (cx, cy+vertical_space), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
        cv.putText(printedImage, f'perimeter {perimeter}', (cx, cy+2*vertical_space), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
        cv.putText(printedImage, f'roundness {roundness}', (cx, cy+3*vertical_space), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
        

# Convert the image from BGR to RGB for Matplotlib
# image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

cv.imshow("printed image", printedImage)
cv.waitKey(0)
```
